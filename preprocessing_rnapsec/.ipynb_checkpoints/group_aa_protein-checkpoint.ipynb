```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_chin = pd.read_csv("../data/chin_all_col.csv", index_col=False)
df_prepro = pd.read_csv("../data/prepro_all_col.csv", index_col=False)
```

```python
ml_chin = pd.read_csv("../data/chin.csv", index_col=False).dropna()
ml_prepro = pd.read_csv("../data/prepro.csv", index_col=False).dropna()
```

```python
ml_chin.dropna()
```

```python
df_chin_2 = df_chin[df_chin.pmid_label.notna()]
df_chin_2 = df_chin_2.rename(columns = {"level_0": "sec_idx"})
df_chin_2 = df_chin_2.reset_index(drop = False)
```

```python
df_chin_2
```

```python
#タンパク質名切り出し、二種以上の融合タンパク質はfusionで統一
df_chin_2["protein_name_seq"] =  df_chin_2.protein_sequence.str.split("|", 2, expand = True)[2].str.split("(\s)", 1, expand = True)[0]
df_chin_2.__getitem__("protein_name_seq").__setitem__(df_chin_2[df_chin_2.protein_name_seq.str.contains("residue", na = False)].index, "Fusion")
df_chin_2["protein_name_2"] = df_chin_2.protein_name_seq.str.replace("(_[A-Z]*)", "")                                          
```

```python
df_chin_2.protein_name_seq[df_chin_2.protein_name_seq.isna()] = "repeat"
```

```python
df_chin_2[df_chin_2.protein_name_seq.isna()].protein_name_seq
```

```python
df_prepro["protein_name_seq"] =  df_prepro.protein_sequence.str.split("|", 2, expand = True)[2].str.split("(\s)", 1, expand = True)[0]
df_prepro.__getitem__("protein_name_seq").__setitem__(df_prepro[df_prepro.protein_name_seq.str.contains("residue", na = False)].index, "Fusion")
df_prepro["protein_name_2"] = df_prepro.protein_name_seq.str.replace("(_[A-Z]*)", "")                                          
```

```python
df_prepro.protein_name_seq[df_prepro.protein_name_seq.isna()] = "repeat"
```

```python
df_prepro[~df_prepro.protein_name_seq.isin(df_chin_2.protein_name_seq.unique())].protein_sequence
```

```python
df_prepro_2 = df_prepro[df_prepro.group_label.notna()].reset_index(drop = False)
```

```python
ml_prepro["protein_name_seq"] = df_prepro_2.protein_name_seq
ml_prepro["aa"] = df_prepro_2.aa
```

```python
ml_chin["protein_name_seq"] = df_chin_2.protein_name_seq
ml_chin["aa"] = df_chin_2.aa
ml_chin["ds_name"] = "chin"
ml_prepro["ds_name"] = "prepro"
```

```python
ml = pd.concat([ml_chin, ml_prepro], axis = "index")
```

```python
from sklearn import preprocessing


def labeling(df, label="protein_name_seq"):
    le = preprocessing.LabelEncoder()
    labels = df[label].unique()
    labels_id = le.fit_transform(df[label])
    df[f"{label}_label"] = labels_id 
    group_label = df[f"{label}_label"]
    return df, group_label
```

```python
ml, df_label = labeling(ml)
ml, df_aa_label = labeling(ml, label = "aa")
```

```python
ml_chin = ml[ml.ds_name == "chin"]
ml_prepro = ml[ml.ds_name == "prepro"]
```

```python
ml_chin[~ml_chin.aa_label.isin(ml_prepro.aa_label)]
```

```python

ml_chin_3 = ml_chin.drop(["aa","protein_name_seq", "ds_name"], axis = "columns")
ml_chin_3.drop(["group_label", "aa_label"], axis = "columns").to_csv("../data/chin_protein_name.csv", index = False)
ml_chin_3.drop(["group_label", "protein_name_seq_label"], axis = "columns").to_csv("../data/chin_aa.csv", index = False)
```

```python

ml_prepro_3 = ml_prepro.drop(["aa","protein_name_seq", "ds_name"], axis = "columns")
ml_prepro_3.drop(["group_label", "aa_label"], axis = "columns").to_csv("../data/prepro_protein_name.csv", index = False)
ml_prepro_3.drop(["group_label", "protein_name_seq_label"], axis = "columns").to_csv("../data/prepro_aa.csv", index = False)
```
